```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.1, 10, 100)
y_sin = np.sin(x)
y_log = np.log(x)
y_exp = np.exp(x)
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.plot(x, y_sin, label='sin(x)')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.subplot(1, 3, 2)
plt.plot(x, y_log, label='log(x)', color='red')
plt.title('Logarithm Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.legend()
plt.subplot(1, 3, 3)
plt.plot(x, y_exp, label='exp(x)', color='red')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.legend()
plt.tight_layout()
plt.show()
```
